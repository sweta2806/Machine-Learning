```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
x = [1,2,3,4,5,6]
y = [11,13,18, 19, 21,24]
```

```python
plt.plot(x,y)
plt.show()
```

```python
x = np.arange(0,10,0.1)
y = np.sin(x)
```

```python
plt.plot(x,y)
plt.show()
```

```python
area = [600,800,1000,1200,1500]
price = [15,18,25,35,40]
```

```python
plt.figure(figsize=(6,4))
plt.plot(area,price)
plt.title("Area vs Price")
plt.xlabel("Area in sq ft")
plt.ylabel("Price in lakh")
plt.xlim(200, 2000)
plt.ylim(5,50)
plt.show()
```

```python
plt.figure(figsize=(7,7))
plt.plot(area,price)
plt.title("Area vs Price")
plt.xlabel("Area in sq ft")
plt.ylabel("Price in lakh")
plt.xticks(rotation=90)
plt.yticks(rotation=270)
plt.show()
```

```python
x = np.arange(1,20)
y1 = np.sqrt(x)
y2 = np.log(x)
```

```python
plt.figure(figsize=(7,7))
plt.plot(x,y1, color = "red")
plt.plot(x,y2, color = "blue")
plt.legend(labels=["square root","log"])
plt.show()
```

# Histograms

```python
df = pd.read_csv("cars.csv")
```

```python
df
```

```python
df.info()
```

```python
df.describe()
```

```python
plt.figure(figsize=(7,7))
```

```python
plt.figure(figsize=(7,7))
plt.hist(df["width"], bins=5)
plt.xticks(rotation=90)
plt.yticks(rotation=45)

plt.title("Width frequency distribution")
plt.ylabel("frequency")
plt.xlabel("width")
plt.show()
```

```python
width_mean = df["width"].mean()
width_median = df["width"].median()
plt.figure(figsize=(7,7))
df["width"].hist(bins=5)
plt.title("Width frequency distribution")
plt.axvline(width_mean,color="red")
plt.axvline(width_median,color="indigo")
plt.legend(labels=["mean","median"])
plt.show()
```

```python
df["width"].hist()
plt.xlabel("width")
plt.ylabel("frequency")
plt.show()
```

# Bar Plot

```python
df.head()
```

```python
df["make"].value_counts().plot(kind="bar")
plt.show()
```

```python
# horizontal plot
plt.figure(figsize=(9,7))
df["make"].value_counts().plot(kind="barh")
plt.title("Make count bar plot")
plt.ylabel("Make")
plt.xlabel("Count")
plt.show()
```

```python
df["body-style"].value_counts().plot.pie(autopct="%1.1f%%", explode=(0,0,0,0.1,0))
plt.title("Body-style count pie chart")
```

```python
df.corr()
```

# Scatter plot

```python
plt.figure(figsize=(7,7))
plt.scatter(x=df["width"],y=df["price"])
plt.show()
```

```python
plt.figure(figsize=(7,7))
plt.scatter(x=df["highway-mpg"],y=df["price"], color="red")
plt.title("City-mpg vs Price Scatter plot")
plt.ylabel("price")
plt.xlabel("city-mpg")
plt.show()
```
